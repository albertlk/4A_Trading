```python
def getFinanceTab(ticker, startDate = "2020-06-01", endDate = "2021-01-15"):
    
    tickerInd = getTechIndicators(getOHLCV(ticker, startDate, endDate), dropnull = False)
    tickerSig = indAnalysis(tickerInd)
    tickerFund = calcFRatios(ticker)
    
    # Plotting prices, SMAs, and oscillators
    closingSMAPlot = tickerInd.hvplot.line(
        xlabel = "Date",
        ylabel = "Price",
        y = ["close", "SMA20", "SMA50", "SMA100"],
        title = f"{ticker} Closing Price Movement",
        shared_axes = False
    )
 
    chaikinPlot = tickerInd.hvplot.line(
        xlabel = "Date",
        ylabel = "CMF",
        y = "CMF",
        title = "Chaikin Money Flow",
        shared_axes = False
    )
         
    rsi_df = pd.DataFrame({
        "RSI":tickerInd["RSI"],
        "Over Bought": 70,
        "Over Sold": 30
    }, index = tickerInd.index)
    
    rsiPlot = rsi_df.hvplot.line(
        xlabel = "Date",
        ylabel = "RSI",
        y = ["RSI", "Over Bought", "Over Sold"],
        title = "Relative Strength Index",
        shared_axes = False
    )
    
    priceCol = pn.Column(f"##Ticker: {ticker}", closingSMAPlot, chaikinPlot, rsiPlot)
                         
    ohlcv_df = getLastOHLCV(tickerSig)
    ohlcv_fig = ohlcv_df.hvplot.table()    
                         
    sig_df = getLastSig(tickerSig)
    sig_fig = sig_df.hvplot.table()
    
    fund_df = getLastFRatio(tickerFund)
    fund_fig = fund_df.hvplot.table()
                         
    indCol = pn.Column(f"###Latest Period: {endDate}", "Candle Data", ohlcv_fig, "Technical Outlook", sig_fig, "Fundamental Ratios", fund_fig)
    
    finTab = pn.Row(priceCol, indCol)
    
    return finTab
```
